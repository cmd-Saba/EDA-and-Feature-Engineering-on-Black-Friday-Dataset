```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Re-load original data to ensure Gender and other columns are in their initial state
df_train_orig = pd.read_csv('train.csv')
df_test_orig = pd.read_csv('test.csv')

# Merge both train and test data to create a clean 'df' for preprocessing
df = pd.concat([df_train_orig, df_test_orig], ignore_index=True)

# Drop User_ID
df.drop(['User_ID'], axis=1, inplace=True)

# Handling Categorical Feature (Gender)
le_gender = LabelEncoder()
df['Gender'] = le_gender.fit_transform(df['Gender'])

# Handling Categorical Feature (Age)
le_age = LabelEncoder()
df['Age'] = le_age.fit_transform(df['Age'])

# Fixing Categorical City_category (one-hot encoding)
df_city_encoded = pd.get_dummies(df['City_Category'], drop_first=True, dtype=int)
df = pd.concat([df, df_city_encoded], axis=1)
df.drop('City_Category', axis=1, inplace=True)

# Handling Missing Values for Product_Category_2 and Product_Category_3
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])

# Handling Stay_In_Current_City_Years
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+', '', regex=False).astype(int)

# Drop Product_ID as it's typically not used in models
df.drop('Product_ID', axis=1, inplace=True)

# Separate into training and prediction sets based on 'Purchase' (NaNs)
df_train_final = df[~df['Purchase'].isnull()]
# df_test_final for actual test set prediction later

# Define X (features) and y (target) for the train_test_split
X = df_train_final.drop('Purchase', axis=1)
y = df_train_final['Purchase']

print("X and y have been defined.")
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```
